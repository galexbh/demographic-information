```python
df = pd.read_csv("/content/drive/MyDrive/UNAH/IA/unemployment analysis.csv", delimiter=",")
df_fertility = pd.read_csv("/content/drive/MyDrive/UNAH/IA/homework/Fertility.csv").drop(columns=["Country"])
df_GDP = pd.read_csv("/content/drive/MyDrive/UNAH/IA/homework/GDP per capita.csv").drop(columns=["Country"])
df_suicide = pd.read_csv("/content/drive/MyDrive/UNAH/IA/homework/Suicide rate.csv").drop(columns=["Country"])
df_population = pd.read_csv("/content/drive/MyDrive/UNAH/IA/homework/Population growth.csv").drop(columns=["Country"])
df_country = df_fertility.merge(df_GDP, how="inner", on="ISO-code")
df_country = df_country.merge(df_suicide, how="inner", on="ISO-code")
df_country = df_country.merge(df_population, how="inner", on="ISO-code")
```

```python
dffinal = pd.DataFrame()
for y in df.columns[ 2 :  ]:
    dftemp  = df.loc[ :  , ['Country Name','Country Code', y] ]
    dftemp['Year'] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dffinal = pd.concat([dffinal, dftemp])
```
